```python
# импорт библиотек
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from google.colab import files
from io import BytesIO
from PIL import Image
```

```python
tf.keras.applications.VGG16
tf.keras.applications.VGG19
```

```python
# указание параметров нейронной сети
tf.keras.applications.VGG16(
 include_top=True,    # использование полносвязных слоёв
 weights="imagenet",    # определение значений весов
 input_tensor=None,
 input_shape=None,
 pooling=None,
 classes=1000,    # определение выходных нейронов
 classifier_activation="softmax",    # функция активации для выходных нейронов
)
```

```python
model = keras.applications.VGG16()

uploaded = files.upload()
img = Image.open(BytesIO(uploaded[list(uploaded.keys())[0]]))
plt.imshow( img )

# приводим к входному формату VGG-сети
img = np.array(img)
x = keras.applications.vgg16.preprocess_input(img)
print(x.shape)
x = np.expand_dims(x, axis=0)
```

```python
 # прогоняем через сеть
res = model.predict( x )
print(np.argmax(res))
```
